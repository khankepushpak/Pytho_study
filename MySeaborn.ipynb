## Distribution Plots

### a. distribution plot
<li> Provides a way to look at a univariate distribution. </li>
<li> A univeriate distribution provides a distribution for one variable</li>
<li> Kernal Density Estimation with a Histogram is provided</li>
<li> kde=False removes the KDE</li>
<li> Bins define how many buckets to divide the data up into between intervals</li>
<li> For example put all profits between $10 and $20 in this bucket</li>

```python
df_real_estate = pd.read_csv("histogram_data.csv")
df_real_estate
```

```python
sns.distplot(df_real_estate["Price"], 
             bins=10,
            kde = True,
             color = 'b')

plt.show()
```

## Kde Plot

```python
# Get just the KDE plot
sns.kdeplot(df_real_estate["Price"])
plt.show()
```

#### Boxplot

<li> A box plot allows you to compare different variables
<li> The box shows the quartiles of the data. The bar in the middle is the median and
<li> the box extends 1 standard deviation from the median
<li> The whiskers extend to all the other data aside from the points that are considered to be outliers

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='day',y='total_bill', data=tips_df, 
            hue='time'
           )
```

```python
# Hue can add another category being sex
# We see men spend way more on Friday versus less than women on Saturday
plt.figure(figsize=(10,6))
plt.yticks = ([5,10,15,20,25,30,35,40,45,50])
sns.boxplot(x='day',y='total_bill',data=tips_df, 
            hue='sex'
           )
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips_df)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips_df)
plt.ylabel('tip ($)')
plt.show()
```
